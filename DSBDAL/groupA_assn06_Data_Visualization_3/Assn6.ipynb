# Boxplot

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

sns.boxplot(ax=axes[0, 0], data=data, x="Species", y="sepal_length")
sns.boxplot(ax=axes[0, 1], data=data, x='Species', y='sepal_width')
sns.boxplot(ax=axes[1, 0], data=data, x='Species', y='petal_length')
sns.boxplot(ax=axes[1, 1], data=data, x='Species', y='petal_width')
```

```python
def find_outlier(arr1):
    q1 = np.quantile(arr1, 0.25)
    q3 = np.quantile(arr1, 0.75)
    med = np.median(arr1)
    
    iqr = q3-q1

    lower_bound = q1 - (1.5*iqr)
    upper_bound = q3 + (1.5*iqr)
    print("IQR= " + str(iqr)  + "\nlower_bound= " + str(lower_bound) + "\nupper_bound= " + str(upper_bound))
    return lower_bound, upper_bound
    
    
def disp_outlier(arr1, lower_bound, upper_bound):
    outliers = arr1[(arr1 <= lower_bound) | (arr1 >= upper_bound)]
    print('The following are the outliers in the boxplot:{}'.format(outliers))
```

```python
lower_bound, upper_bound = find_outlier(data["sepal_length"])
```
