The objective is to predict the value of prices of the house using the given features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 data = sns.load_dataset('titanic')
```

```python
data.info()
```

```python
data
```

```python
sns.displot(data, x="age", col="sex")
```

```python
sns.countplot(x='survived', hue='pclass', data=data)
```

```python
sns.countplot(x='survived', hue='sex', data=data)
```

```python
sns.barplot(data=data, x='pclass', y='sex', hue='survived')
```

```python
sns.distplot(data['fare'], bins=8)
```

```python
data.skew()
```

```python
sns.swarmplot(data=data, x='fare')
```

```python
sns.swarmplot(data=data, x='age')
```

```python
sns.swarmplot(data=data, x='survived', y='fare')
```

```python
sns.boxplot(x=data["age"])
```

```python
sns.boxplot(data=data, x="sex", y="age", hue="survived")
```

```python
sns.boxplot(data=data, x="sex", y="age")
```
